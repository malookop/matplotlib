Advance Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://github.com/YBI-foundation/Dataset/raw/main/Tips.csv")
```

```python
df.head()
```

making graph from dataframe

ax.scatter("col1","col2",data=dataframe_name)

1 way to use dataframe---seaborn also use this way

```python
fig,ax=plt.subplots()
ax.scatter("total_bill","tip",data=df)
```

defing with dataframe

2 way of using dataframe

```python
fig,ax=plt.subplots()
ax.scatter(df["total_bill"],df["tip"])
```

```python
fig,ax=plt.subplots()
ax.bar("day","total_bill",data=df)
```

```python
fig,ax=plt.subplots()
ax.barh(df["day"],df['total_bill'])
```

```python
fig,ax=plt.subplots()
ax.pie("size",data=df);
```

visualisaton in more subplots

ax[address],address=row,col

```python
fig,ax=plt.subplots(nrows=2,ncols=2)
ax[1,1].scatter('total_bill',"tip",data=df)
ax[0,1].bar('day',"total_bill",data=df)


```

note=data can be from diffrent dataframe or tables

multiple graph or chart can be used in single sublot

```python
fig,ax=plt.subplots(nrows=2,ncols=2)
ax[1,1].scatter('total_bill',"tip",data=df)
ax[1,1].bar('total_bill',"tip",data=df)
ax[0,1].bar('day',"total_bill",data=df)
ax[0,0].barh("day","total_bill",data=df)
ax[1,0].pie(df["size"]);
```
